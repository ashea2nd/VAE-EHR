```python
from rpy2.robjects import r, pandas2ri
import rpy2.robjects as robjects
from rpy2.robjects import default_converter
from rpy2.robjects.conversion import localconverter
import rpy2
pandas2ri.activate()

ehrfeat_path = "/Users/andrew/Documents/meng/mixEHR/data/Mimic/version_1_3/processed/ehrFeatId.RData"
ehrfeat = r.load(ehrfeat_path)

ehrfeatinfo = robjects.globalenv.get(ehrfeat[0])
print(len(ehrfeat))
print("Type:", type(ehrfeat))
print("Environment Keys:", tuple(robjects.globalenv.keys()))
print("ehrfeatinfo:", dir(ehrfeatinfo))
print("icd_cm index:", ehrfeatinfo.names.index("icd_cm"))
print("icd_cpt index:", ehrfeatinfo.names.index("icd_cpt"))

icd_cm_df = ehrfeatinfo[1]
icd_cpt_df = ehrfeatinfo[2]
print(dir(icd_cm_df))
```

```python
icdcm_typeId_idx = icd_cm_df.colnames.index("typeId")
icdcm_pheId_idx = icd_cm_df.colnames.index("pheId")
icdcm_ICD9_idx = icd_cm_df.colnames.index("ICD9_CODE")


icdcm_typeId_col = icd_cm_df[icdcm_typeId_idx]
icdcm_pheId_col = icd_cm_df[icdcm_pheId_idx]

icdcm_ICD9_fv = icd_cm_df[icdcm_ICD9_idx] #factor vector
icdcm_ICD9_levels = icdcm_ICD9_fv.levels
icdcm_ICD9_col = [icdcm_ICD9_levels[i - 1] for i in icdcm_ICD9_fv]
icdcm_ICD9_col
```

```python
icdcpt_typeId_idx = icd_cpt_df.colnames.index("typeId")
icdcpt_pheId_idx = icd_cpt_df.colnames.index("pheId")
icdcpt_ICD9_idx = icd_cpt_df.colnames.index("ICD9_CODE")

icdcpt_typeId_col = icd_cpt_df[icdcpt_typeId_idx]
icdcpt_pheId_col = icd_cpt_df[icdcpt_pheId_idx]
icdcpt_ICD9_fv = icd_cpt_df[icdcpt_ICD9_idx] #factor vector

icdcpt_ICD9_levels = icdcpt_ICD9_fv.levels
icdcpt_ICD9_col = [icdcpt_ICD9_levels[i - 1] for i in icdcpt_ICD9_fv]
icdcpt_ICD9_col
```

```python
icdcm_df = pd.DataFrame(list(zip(icdcm_typeId_col, icdcm_pheId_col, icdcm_ICD9_col)), columns=["typeId", "pheId", "ICD9_CODE"])
icdcm_df
```

```python
icdcpt_df = pd.DataFrame(list(zip(icdcpt_typeId_col, icdcpt_pheId_col, icdcpt_ICD9_col)), columns=["typeId", "pheId", "ICD9_CODE"])
icdcpt_df
```

```python
icd_df = icdcm_df.append(icdcpt_df, ignore_index=True)
icd_df
```

```python
phi_with_pheNames = pd.read_csv("/Users/andrew/Documents/meng/mixEHR/data/precomputed_results/mixmimic/mimic_trainData_JCVB0_nmar_K75_iter5497_phi_normalized_WITH_PHENAMES.csv")
phi_with_pheNames
```

```python
phi_with_icd_df = pd.merge(phi_with_pheNames, icd_df, on=["typeId", "pheId"])
phi_with_icd_df
```

```python
icd9_code_col = phi_with_icd_df["ICD9_CODE"]

phi_with_icd_df = phi_with_icd_df.drop(labels=["ICD9_CODE"], axis=1)
phi_with_icd_df.insert(3, "ICD9_CODE", icd9_code_col)
phi_with_icd_df
```

```python
phi_with_icd_df.to_csv("mimic_trainData_JCVB0_nmar_K75_iter5497_phi_normalized_WITH_PHENAMES_ICD.csv", index=False)
```
