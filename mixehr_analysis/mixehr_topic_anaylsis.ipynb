```python
phi_path = "/Users/andrew/Documents/meng/mixEHR/data/precomputed_results/mixmimic/mimic_trainData_JCVB0_nmar_K75_iter5497_phi_normalized.csv"
phi = pd.read_csv(phi_path, header=None)
print(phi.iloc[:, 0].unique())
print("Unique pheIds", len(phi.iloc[:, 1].unique()))
print(phi.groupby([0, 1]).size())
phi
```

```python
from rpy2.robjects import r, pandas2ri
import rpy2.robjects as robjects
from rpy2.robjects import default_converter
from rpy2.robjects.conversion import localconverter
import rpy2
pandas2ri.activate()

ehrfeat_path = "/Users/andrew/Documents/meng/mixEHR/data/Mimic/version_1_3/processed/ehrFeatId.RData"
ehrfeat = r.load(ehrfeat_path)
print(len(ehrfeat))
print("Type", type(ehrfeat))
print("Environment Keys", tuple(robjects.globalenv.keys()))

ehrfeatinfo = robjects.globalenv.get(ehrfeat[0])

pheNames = []
phi_typeId_pheId = phi.iloc[:, [0,1]]
for idx, row in tqdm(phi_typeId_pheId.iterrows()):
    typeId = row.iloc[0]
    pheId = row.iloc[1]    
    typeId_df = ehrfeatinfo[typeId-1]
    pheName_vector = typeId_df.rx(typeId_df.rx('pheId').ro == pheId, True).rx('pheName')[0]
    if type(pheName_vector) == rpy2.robjects.vectors.FactorVector:
        pheName_factorlevel = pheName_vector[0]
        pheName = pheName_vector.levels[pheName_factorlevel - 1]
        pheNames.append(pheName)
    elif type(pheName_vector) == rpy2.robjects.vectors.StrVector:
        pheName = list(pheName_vector)[0]
        pheNames.append(pheName)
```

```python
print(pheNames)
print(len(pheNames))
```

```python
phi["pheName"] = pheNames
phi
```

```python
phi.to_csv("mimic_trainData_JCVB0_nmar_K75_iter5497_phi_normalized_WITH_PHENAMES.csv", index=False)
```
